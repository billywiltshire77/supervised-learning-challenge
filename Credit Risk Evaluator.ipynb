```python
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
```

```python
train_df.columns
```

```python
train_df['loan_status'].value_counts()
```

```python
# Convert categorical data to numeric and separate target feature for training data
y_train = LabelEncoder().fit_transform(train_df['loan_status'])
X_train = train_df.drop('loan_status', axis=1)
y_train
```

```python
X_train_dummies = pd.get_dummies(X_train)
X_train_dummies.head()
```

```python
# Convert categorical data to numeric and separate target feature for testing data
y_test = LabelEncoder().fit_transform(test_df['loan_status'])
X_test = test_df.drop('loan_status', axis=1)
```

```python
X_test_dummies = pd.get_dummies(X_test)
X_test_dummies.head()
```

```python
# add missing dummy variables to testing set
for column in X_train_dummies.columns:
    if column not in X_test_dummies.columns:
        print(column)
```

```python
debt_settlement_flag_Y = []
for value in X_test_dummies['debt_settlement_flag_N']:
    if value == 0:
        debt_settlement_flag_Y.append(1)
    else:
        debt_settlement_flag_Y.append(0)
```

```python
X_test_dummies['debt_settlement_flag_Y'] = debt_settlement_flag_Y
X_test_dummies.head()
```

Pre Analysis Prediction:

Before analyzing the data using a logistic regression and random forest classifier, I believe the better model will be the logistic regression. I believe this because there are too many factors involving loan approval and credit risk (94 features) with overlap. The random forest classifier will have a hard time classifying without over fitting whereas the logistic regression will be able to assign weighting to each feature and predict with greater accuracy. I believe that both scores will improve with standardization as the data will be on the same scale.

```python
# Train the Logistic Regression model on the unscaled data and print the model score
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train_dummies, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train_dummies, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test_dummies, y_test)}")
```

```python
# Train a Random Forest Classifier model and print the model score
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(random_state=1, n_estimators=500).fit(X_train_dummies, y_train)
```

```python
print(f'Training Score: {clf.score(X_train_dummies, y_train)}')
print(f'Testing Score: {clf.score(X_test_dummies, y_test)}')
```

```python
# Scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(X_train_dummies)
X_train_scaled = scaler.transform(X_train_dummies)
X_test_scaled = scaler.transform(X_test_dummies)
```

```python
# Train the Logistic Regression model on the scaled data and print the model score
classifier.fit(X_train_scaled, y_train)
print(f"Training Data Score: {classifier.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test_scaled, y_test)}")
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score
clf = RandomForestClassifier(random_state=1, n_estimators=500).fit(X_train_scaled, y_train)
print(f'Training Score: {clf.score(X_train_scaled, y_train)}')
print(f'Testing Score: {clf.score(X_test_scaled, y_test)}')
```

Supervised Learning Analysis:
After performing a logistic regression and random forest classifier analysis on both non scaled and scaled data,the results showed that it was hard to predict to a very high degreee of certainty if someone was worthy of being approved for a loan. The model that did the best was the logistic regression analysis using the scaled data which imporved significantly over the non scaled. It is intresting that the random forest classifier did not significantly improve after scaling the data after out performing the logistic regression on the non scaled data. From my original prediction, I was right and wrong. I was right that the logistic regression would be better at predicting but that was only after scaling the data. Similarly, scaling improved the logistic regression score but failed to improve the random forest classifier. I believe this is due to the nature of finding branches in the decision tree and then averaging the results. Standardizing the data wouldn't change the decisions the model would make so the fact it did not improve makes sense.
